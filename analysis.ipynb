## Data processing

We are only going to focus on data from 2013-2014 season beyond, so injury data older than 2013 will be ignored.

```python
print(f'Number of rows: {df_injuries.shape[0]}')
regex_year = '2013-1[0-2]-(0[1-9]|[1-2][0-9]|3[0-1])|(201[4-9]|2020)-(0[1-9]|1[0-2])-(0[1-9]|[1-2][0-9]|3[0-1])'
df_injuries_processed = df_injuries[df_injuries['Date'].str.contains(regex_year)]
print(f'Number of rows: {df_injuries_processed.shape[0]}')
df_injuries_processed.head()
```

```python
def get_df_player_data_season(season,league_id):
    
    defend_data = leaguedashptdefend.LeagueDashPtDefend(
                season = season,
                per_mode_simple='PerGame',
                defense_category='Overall',
                league_id = league_id,
                season_type_all_star = 'Regular Season').get_data_frames()[0]
    
    bio_data = leaguedashplayerbiostats.LeagueDashPlayerBioStats(
                season = season,
                per_mode_simple='PerGame',
                league_id = league_id,
                season_type_all_star= 'Regular Season').get_data_frames()[0]
                
    df_players = defend_data.copy()

    df_players = df_players.drop(columns = ['FREQ','G','PCT_PLUSMINUS','NORMAL_FG_PCT'])
    df_players = df_players.rename(columns = {'CLOSE_DEF_PERSON_ID':'PLAYER_ID'})

    cols_to_merge = ['PLAYER_ID','PLAYER_HEIGHT','PLAYER_HEIGHT_INCHES','PLAYER_WEIGHT']

    df_players = df_players.merge(bio_data[cols_to_merge])
    
    return df_players
    
```
