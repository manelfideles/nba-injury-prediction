## Data Aggreagtion

```python
def get_df_player_data_season(season,league_id,season_segment):
    
    defend_data = leaguedashptdefend.LeagueDashPtDefend(
                season = season,
                per_mode_simple='PerGame',
                defense_category='Overall',
                league_id = league_id,
                season_type_all_star = season_segment).get_data_frames()[0]

    speed_distance_data = leaguedashptstats.LeagueDashPtStats(
                            season = season,
                            season_type_all_star = season_segment,
                            per_mode_simple = 'PerGame',
                            player_or_team = 'Player',
                            # month = ,
                            ).get_data_frames()[0]

    bio_data = leaguedashplayerbiostats.LeagueDashPlayerBioStats(
                season = season,
                per_mode_simple='PerGame',
                league_id = league_id,
                season_type_all_star= season_segment).get_data_frames()[0]

    touches_data = 0
    
    hustle_data = leaguehustlestatsplayer.LeagueHustleStatsPlayer(
                            season = '2019-20',
                            season_type_all_star = 'Regular Season',
                            per_mode_time = 'PerGame',
                            # month = ,
                            ).get_data_frames()[0]

    df_players = defend_data.copy()

    df_players = df_players.drop(columns = ['FREQ','G','PCT_PLUSMINUS','NORMAL_FG_PCT'])
    df_players = df_players.rename(columns = {'CLOSE_DEF_PERSON_ID':'PLAYER_ID','PLAYER_NAME':'name','PLAYER_POSITION':'position'})

    cols_to_merge = ['PLAYER_ID','PLAYER_HEIGHT','PLAYER_HEIGHT_INCHES','PLAYER_WEIGHT']

    df_players = df_players.merge(bio_data[cols_to_merge])
    
    cols_to_merge = ['PLAYER_ID','DIST_FEET','DIST_MILES','DIST_MILES_OFF','DIST_MILES_DEF','AVG_SPEED','AVG_SPEED_OFF','AVG_SPEED_DEF']

    df_players = df_players.merge(speed_distance_data[cols_to_merge])

    cols_to_merge = ['PLAYER_ID','CONTESTED_SHOTS','CONTESTED_SHOTS_2PT','CONTESTED_SHOTS_3PT','BOX_OUTS','OFF_BOXOUTS','DEF_BOXOUTS']

    df_players = df_players.merge(hustle_data[cols_to_merge])
    
    df_players = df_players.drop(columns = ['PLAYER_HEIGHT'])
    df_players = df_players.rename(columns ={'PLAYER_ID':'id','PLAYER_HEIGHT':'height', 'PLAYER_HEIGHT_INCHES': 'inches','PLAYER_WEIGHT':'weight'} )
    return df_players
```

#### Player Dataset Glossary
 - GP: Games played in segment
 - D_FGM: Defensive field goals made, like rebounds
 - D_FGA: Defensive field goals attemped
 - D_FG_PCT: Defensive field goals percentage
 - DIST_FEET: Total distance in feet made
 - DIST_MILES: Total distance in miles made
 - AVG_SPEED: speed of how many miles the player can run per second
 - inches: height in inches, which can be used to calculate in feet or cm

```python
df_players = get_df_player_data_season('2019-20','00','Regular Season')
df_players
```
