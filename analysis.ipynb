## Data Processing

### Historic Data

Trying to create a automated way to get the historic data so we can analyse the performance of a player before injury

**Observations**
- Some endpoints have **month** parameter others don't
- Difference of weight can be a indicator of injury
    - This measure will only be year difference and not month difference beacuse players during the season focus on mantaining their physic rather than growing. They build more mass between the end of the season and the start of the next 
- Difference in height won't make a difference, most of the players maintain their height since it's a biological thing and its out of the control of athletes
- Difficulty in joining data in a interval of time beacuse new players come in every season as well some retire
- Defense 
    - Endpoint: 
- Speed & Distance 
    - Endpoint: https://github.com/swar/nba_api/blob/master/docs/nba_api/stats/endpoints/leaguedashptstats.md
- Hustle
    - Endpoint:
- Touches
- Drive


```python
dict_season = {i: f'{i}-{(i%2000)+1}'for i in range(2013,2023)}
columns_sum = ['DIST_FEET','DIST_MILES','DIST_MILES_OFF','DIST_MILES_DEF']
columns_mean = ['CONTESTED_SHOTS','CONTESTED_SHOTS_2PT','CONTESTED_SHOTS_3PT','BOX_OUTS','OFF_BOXOUTS',
                    'DEF_BOXOUTS','DEFLECTIONS','CHARGES_DRAWN','AVG_SPEED','AVG_SPEED_OFF','AVG_SPEED_DEF']

def get_player_season_data(season,league_id,season_segment,month = None):

    defend_data = leaguedashptdefend.LeagueDashPtDefend(
                season = season,
                per_mode_simple='PerGame',
                defense_category='Overall',
                league_id = league_id,
                season_type_all_star = season_segment).get_data_frames()[0]

    speed_distance_data = leaguedashptstats.LeagueDashPtStats(
                            season = season,
                            season_type_all_star = season_segment,
                            per_mode_simple = 'PerGame',
                            player_or_team = 'Player',
                            ).get_data_frames()[0]
    
    touches_data = 0
    
    hustle_data = leaguehustlestatsplayer.LeagueHustleStatsPlayer(
                            season = season,
                            season_type_all_star = 'Regular Season',
                            league_id_nullable = league_id,
                            per_mode_time = 'PerGame',
                            ).get_data_frames()[0]

    df_players = defend_data.copy()

    df_players = df_players.drop(columns = ['FREQ','G','PCT_PLUSMINUS','NORMAL_FG_PCT'])
    df_players = df_players.rename(columns = {'CLOSE_DEF_PERSON_ID':'PLAYER_ID'})

    cols_to_merge = ['PLAYER_ID','DIST_FEET','DIST_MILES','DIST_MILES_OFF','DIST_MILES_DEF','AVG_SPEED','AVG_SPEED_OFF','AVG_SPEED_DEF']

    df_players = df_players.merge(speed_distance_data[cols_to_merge])

    cols_to_merge = ['PLAYER_ID','CONTESTED_SHOTS','CONTESTED_SHOTS_2PT','CONTESTED_SHOTS_3PT','BOX_OUTS','OFF_BOXOUTS','DEF_BOXOUTS','DEFLECTIONS','CHARGES_DRAWN']

    df_players = df_players.merge(hustle_data[cols_to_merge])
    return df_players


def get_player_historic(start,end,league_id,season_segment):
    
    day_s,month_s,year_s = start.split("/")

    day_e,month_e,year_e = end.split("/")
    
    year_s = int(year_s)
    year_e = int(year_e)
    month_s = int(month_s)
    month_e = int(month_e)
    day_s = int(day_s)
    day_e = int(day_e)

    df_historic = leaguedashplayerbiostats.LeagueDashPlayerBioStats(
                season = dict_season[year_s],
                per_mode_simple='PerGame',
                league_id = league_id,
                season_type_all_star= season_segment).get_data_frames()[0]
    
    if year_s < year_e:
        df_bio_end = leaguedashplayerbiostats.LeagueDashPlayerBioStats(
                    season = dict_season[year_e],
                    per_mode_simple='PerGame',
                    league_id = league_id,
                    season_type_all_star= season_segment).get_data_frames()[0]
        
        df_bio_end = df_bio_end.astype({'PLAYER_WEIGHT':int})
        df_historic = df_historic.astype({'PLAYER_WEIGHT':int})

        df_historic['weight_diff'] = df_bio_end['PLAYER_WEIGHT'] - df_historic['PLAYER_WEIGHT']
        # df_historic['weight_end'] = df_bio_end['PLAYER_WEIGHT']

    df_historic = df_historic.drop(columns = ['PLAYER_HEIGHT','COLLEGE','COUNTRY','DRAFT_YEAR','DRAFT_ROUND','DRAFT_NUMBER','OREB_PCT',
                                                    'DREB_PCT','USG_PCT','TS_PCT','AST_PCT','PTS','REB','AST','NET_RATING'])
    

    year_gap = year_e - year_s

    #rest of the stats
    if year_s < year_e:
        for year in range(year_s,year_e):
            df_aux = get_player_season_data(dict_season[year],'00','Regular Season')

            df_historic = df_historic.sort_values(by='PLAYER_ID')
            df_aux = df_aux.sort_values(by='PLAYER_ID')
        
            if set(columns_sum).issubset(df_historic.columns): df_historic[columns_sum] = df_historic[columns_sum] + df_aux[columns_sum]
            else: 
                df_historic = df_historic.merge(df_aux[['PLAYER_ID']+columns_sum])

            print(df_aux[['PLAYER_NAME']+columns_sum].head())
            print(df_historic[['PLAYER_NAME']+columns_sum].head()) 
        
    #     if month_s < year_e:
    #         for i in range(month_s,month_e):
    #             df_aux = get_players_data(dict_season[year],'00','Regular Season')
    #     df_historic[columns_mean] = df_historic[columns_mean]/year_gap
        
    # elif year_s == year_e:
    #     if month_s < year_e:
    #         for i in range(month_s,month_e):
    #             df_historic = get_players_data(dict_season[year],'00','Regular Season')        

    df_historic = df_historic.rename(columns ={'PLAYER_NAME':'name','PLAYER_ID':'id','PLAYER_HEIGHT':'height', 'PLAYER_HEIGHT_INCHES': 'inches','PLAYER_WEIGHT':'weight','PLAYER_POSITION':'position'} )

    return df_historic
```

#### Player Dataset Glossary
 - GP: Games played in segment
 - D_FGM: Defensive field goals made, like rebounds
 - D_FGA: Defensive field goals attemped
 - D_FG_PCT: Defensive field goals percentage
 - DIST_FEET: Total distance in feet made
 - DIST_MILES: Total distance in miles made
 - AVG_SPEED: speed of how many miles the player can run per second
 - inches: height in inches, which can be used to calculate in feet or cm
